```python
import pandas as pd
from django_pandas.io import read_frame
from analysis.models import RuralCommunityRapidAssessment

qs = RuralCommunityRapidAssessment.objects.all()

field_names = [
    'subject_identifier', 'community', 'district', 'gender', 'age_in_years', 'role_other', 'covid19_awareness',
    'covid19_learning_other', 'vaccine_assess', 'vaccine_assess_reason', 'vaccine_assess_challanges',
    'va_challanges_other', 'vaccine_trust', 'vaccine_trust_other', 'vaccine_recommendation',
    'vaccine_recommendation_other', 'vaccination_importance', 'vaccine_reco_others', 'vaccine_recom_other',
    'vaccine_reco_someone', 'vaccine_reco_someone_other', 'role', 'covid19_learning',
]

exclude_fields = [
    'created', 'modified', 'user_created', 'user_modified', 'hostname_created', 'hostname_modified', 'revision',
    'device_created', 'device_modified', 'id', 'site', 'subject_identifier_as_pk', 'subject_identifier_aka',
    'report_date', 'plot_number', 'full_name', 'contracts'
]

#Data dictionary
field_question = []
for field in RuralCommunityRapidAssessment._meta.get_fields():
    if not field.name in exclude_fields:
        field_question.append([field.name, field.verbose_name])
df_dict = pd.DataFrame(field_question)
df_dict.to_csv('RuralCommunityRapidAssessmentDataDict.csv', encoding='utf-8', index=False)

#Export data
df = read_frame(qs, fieldnames=field_names)
df.to_csv('RuralCommunityRapidAssessment.csv', encoding='utf-8', index=False)
```
