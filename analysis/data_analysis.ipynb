```python
import os, sys
import django
from dateutil import parser
from django.utils.timezone import make_aware

sys.path.append('../..') # add path to project root dir
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'flourish.settings')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

# for more sophisticated setups, if you need to change connection settings (e.g. when using django-environ):
os.environ["DATABASE_URL"] = "postgres://myuser:mypassword@localhost:54324/mydb"

# Connect to Django ORM
django.setup()
```

```python
from analysis.models import RuralCommunityRapidAssessment
from pprint import pprint

qs = RuralCommunityRapidAssessment.objects.all()
exclude_fields = [
    'created', 'modified', 'user_created', 'user_modified', 'hostname_created', 'hostname_modified', 'revision',
    'device_created', 'device_modified', 'id', 'site_id', 'subject_identifier_as_pk', 'subject_identifier_aka',
    'report_date', 'plot_number', 'full_name', 'contracts', '_state'
]

data = []
for q in qs:
    role = [t.name for t in q.role.all()]
    covid19_learning = [t.name for t in q.covid19_learning.all()]
    q = q.__dict__
    q.update(role=role)
    q.update(rcovid19_learning=covid19_learning)
    for key in exclude_fields:
      del q[key]
    data.append(q)
df = pd.DataFrame(data)
```

```python
df.to_csv('RuralCommunityRapidAssessment.csv', encoding='utf-8', index=False)
```

```python
#Data dictionary
field_question = []
for field in RuralCommunityRapidAssessment._meta.get_fields():
    if not field.name in exclude_fields:
        field_question.append({'Variable': field.name, 'Question': field.verbose_name})
df_dict = pd.DataFrame(field_question)
```
